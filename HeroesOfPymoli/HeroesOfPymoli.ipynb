```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(10)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uitem=len(purchase_data["Item ID"].unique())
avg_price=purchase_data["Price"].mean()
purchase_amount=purchase_data["Purchase ID"].count()
total_revenue=purchase_data["Price"].sum()

summary=pd.DataFrame({"Number of Unique Items":[uitem],
                      "Average Price":["$"+str("%.2f"%avg_price)],
                      "Number of Purchases":[purchase_amount],
                      "Total Revenue":["$"+str(total_revenue)]})

summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group=purchase_data.groupby(["Gender"])
gender_count=gender_group.count()
male=gender_count.loc["Male","SN"]
female=gender_count.loc["Female","SN"]
other=gender_count.loc["Other / Non-Disclosed","SN"]
total=male+female+other

gender_demo=pd.DataFrame({"Total Count":[male,female,other],
                          "Percentage of Players":[str(round(male/total*100,2))+"%",
                                                   str(round(female/total*100,2))+"%",
                                                   str(round(other/total*100,2))+"%"]})
gender_demo.index=["Male","Female","Other/Non-Disclosed"]

gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_mean=gender_group.mean().round(2)
gender_total=gender_group.sum().round(2)

avgprice_male=str("%.2f"%gender_mean.loc["Male","Price"])
avgprice_female=str("%.2f"%gender_mean.loc["Female","Price"])
avgprice_other=str("%.2f"%gender_mean.loc["Other / Non-Disclosed","Price"])
totalvalue_male=str("%.2f"%gender_total.loc["Male","Price"])
totalvalue_female=str("%.2f"%gender_total.loc["Female","Price"])
totalvalue_other=str("%.2f"%gender_total.loc["Other / Non-Disclosed","Price"])
perpersonvalue_male=str("%.2f"%(float(totalvalue_male)/len(purchase_data["SN"][purchase_data["Gender"]=="Male"].unique())))
perpersonvalue_female=str("%.2f"%(float(totalvalue_female)/len(purchase_data["SN"][purchase_data["Gender"]=="Female"].unique())))
perpersonvalue_other=str("%.2f"%(float(totalvalue_other)/len(purchase_data["SN"][purchase_data["Gender"]=="Other / Non-Disclosed"].unique())))

purchase_gender=pd.DataFrame({"Purchase Count":[male,female,other],
                              "Average Purchase Price":["$"+avgprice_male,
                                                        "$"+avgprice_female,
                                                        "$"+avgprice_other],
                              "Total Purchase Value":["$"+totalvalue_male,
                                                      "$"+totalvalue_female,
                                                      "$"+totalvalue_other],
                              "Avg Total Purchase per Person":["$"+perpersonvalue_male,
                                                               "$"+perpersonvalue_female,
                                                               "$"+perpersonvalue_other]})
purchase_gender.index=["Male","Female","Other/Non-Disclosed"]
purchase_gender.index.name="Gender"

purchase_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_data=purchase_data[["Age","SN","Price"]]
age_unique=age_data.drop_duplicates(subset='SN')
age_bins=[0,9,14,19,24,29,34,39,120]
age_label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
age_unique["Group"]=pd.cut(age_unique["Age"],age_bins,labels=age_label)
age_ugroup=age_unique.groupby("Group")
age_ucount=age_ugroup.count()
age_ucount["Percentage"]=round(100*age_ucount["Age"]/age_ucount["Age"].sum(),2).astype(str)+"%"

age_demo=pd.DataFrame({"Total Count":age_ucount["Age"],
                       "Percentage of Players":age_ucount["Percentage"]})
age_demo.index=age_label

age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_data["Age Ranges"]=pd.cut(age_data["Age"],age_bins,labels=age_label)
age_group=age_data.groupby("Age Ranges")
age_count=age_group["Age"].count()
age_avgprice=age_group["Price"].mean().apply(lambda x: "$"+str("%.2f"%x))
age_totalvalue=age_group["Price"].sum().apply(lambda x: "$"+str("%.2f"%x))
age_perpersonvalue=(age_group["Price"].sum()/age_ucount["SN"]).apply(lambda x: "$"+str("%.2f"%x))

purchase_age=pd.DataFrame({"Purchase Count":age_count,
                           "Average Purchase Price":age_avgprice,
                           "Total Purchase Value":age_totalvalue,
                           "Avg Total Purchase per Person":age_perpersonvalue},
                          index=age_label)

purchase_age.index.name="Age Ranges"

purchase_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spend_data=purchase_data[["SN","Price"]]
spend_data.head()
spend_group=spend_data.groupby("SN")
spend_rank=spend_group.sum().sort_values("Price",ascending=False)
spend_rank["Count"]=spend_group.count()
spend_highrank=spend_rank.iloc[0:5,:]

spend_highrank["Avg Price"]=(spend_highrank["Price"]/spend_highrank["Count"]).apply(lambda x: "$"+str("%.2f"%x))
spend_highrank["Price"]=spend_highrank["Price"].apply(lambda x: "$"+str("%.2f"%x))

top_spenders=pd.DataFrame({"Purchase Count":spend_highrank["Count"],
                           "Average Purchase Price":spend_highrank["Avg Price"],
                           "Total Purchase Value":spend_highrank["Price"]})
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data=purchase_data[["Item ID","Item Name","Price"]]
item_group=item_data.groupby(["Item ID","Item Name"])
item_count=pd.Series(item_group.count()["Price"])
item_total=pd.Series(item_group.sum()["Price"])
popular_item=pd.DataFrame({"Purchase Count":item_count,
                           "Item Price":item_total/item_count,
                           "Total Purchase Value":item_total})
popular_item=popular_item.sort_values("Purchase Count",ascending=False)
popular_item["Item Price"]=popular_item["Item Price"].apply(lambda x: "$"+str("%.2f"%x))
popular_item["Total Purchase Value"]=popular_item["Total Purchase Value"].apply(lambda x: "$"+str("%.2f"%x))
popular_item.head(15)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable=popular_item
most_profitable["Total Purchase Value"]=item_total
most_profitable=most_profitable.sort_values("Total Purchase Value",ascending=False)
most_profitable["Total Purchase Value"]=most_profitable["Total Purchase Value"].apply(lambda x:
                                                                                     "$"+str("%.2f"%x))
most_profitable.head()
```
